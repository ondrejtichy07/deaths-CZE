# Počet úmrtí v ČR v jednotlivých týdnech roku

```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
#import seaborn as sns
```

```python
df = pd.read_csv("https://www.czso.cz/documents/62353418/138258879/130185-20data122220.csv/23f34493-7b02-4f94-8ce8-bf162ee2d860?version=1.1") 
#("/Users/ondrejtichy/Desktop/deaths/130185-20data122220.csv")
```

```python
def make_plot(year):
    f1 = df[df["rok"]==year]
    f2_014 = f1[f1["vek_txt"]=="0-14"]
    f2_1539 = f1[f1["vek_txt"]=="15-39"]
    f2_4064 = f1[f1["vek_txt"]=="40-64"]
    f2_6574 = f1[f1["vek_txt"]=="65-74"]
    f2_85 = f1[f1["vek_txt"]=="85 a více"]

    d = f1[f1.columns[[7,8,12,1]]].sort_values(by='hodnota')
    #print(d[d["vek_txt"] != "celkem"])

    tyden_014 = f2_014[f2_014.columns[[7,8,1]]]["tyden"].tolist()
    hodnota_014 = f2_014[f2_014.columns[[7,8,1]]]["hodnota"].tolist()
    tyden_1539 = f2_1539[f2_1539.columns[[7,8,1]]]["tyden"].tolist()
    hodnota_1539 = f2_1539[f2_1539.columns[[7,8,1]]]["hodnota"].tolist()
    tyden_4064 = f2_4064[f2_4064.columns[[7,8,1]]]["tyden"].tolist()
    hodnota_4064 = f2_4064[f2_4064.columns[[7,8,1]]]["hodnota"].tolist()
    tyden_6574 = f2_6574[f2_6574.columns[[7,8,1]]]["tyden"].tolist()
    hodnota_6574 = f2_6574[f2_6574.columns[[7,8,1]]]["hodnota"].tolist()
    tyden_85 = f2_85[f2_85.columns[[7,8,1]]]["tyden"].tolist()
    hodnota_85 = f2_85[f2_85.columns[[7,8,1]]]["hodnota"].tolist()    

    suma_all = f'{f1[f1["vek_txt"]=="celkem"]["hodnota"].sum():,}'
    suma_014 = f'{f1[f1["vek_txt"]=="0-14"]["hodnota"].sum():,}'
    suma_1539 = f'{f1[f1["vek_txt"]=="15-39"]["hodnota"].sum():,}'
    suma_4064 = f'{f1[f1["vek_txt"]=="40-64"]["hodnota"].sum():,}'
    suma_6574 = f'{f1[f1["vek_txt"]=="65-74"]["hodnota"].sum():,}'
    suma_7584 = f'{f1[f1["vek_txt"]=="75-84"]["hodnota"].sum():,}'
    suma_85 = f'{f1[f1["vek_txt"]=="85 a více"]["hodnota"].sum():,}'

    print(f"Suma celkem : {suma_all}")
    print(f"Suma (0-14) : {suma_1539}")
    print(f"Suma (40-64): {suma_4064}")
    print(f"Suma (65-74): {suma_6574}")
    print(f"Suma (75-84): {suma_7584}")
    print(f"Suma (85+)  : {suma_85}")

    fig, axs = plt.subplots(2, 3, figsize=(15,7))
    axs[-1, -1].axis('off')
  
    fig.suptitle('Počty úmrtí v roce 2020', fontsize=16)
    axs[0, 0].plot(tyden_014,hodnota_014, color="red")
    axs[0, 0].set_title('0-14 let')
    axs[0, 0].set_ylim([0,100])
    axs[0, 0].set_xlim([1,52])
    axs[0, 1].plot(tyden_1539,hodnota_1539, color="red")
    axs[0, 1].set_title('15-39 let')
    axs[0, 1].set_ylim([0,100])
    axs[0, 1].set_xlim([1,52])
    axs[0, 2].plot(tyden_4064,hodnota_4064, color="red")
    axs[0, 2].set_title('40-64 let')
    axs[0, 2].set_ylim([0,1000])
    axs[0, 2].set_xlim([1,52])
    axs[1, 0].plot(tyden_6574,hodnota_6574, color="red")
    axs[1, 0].set_title('65-74 let')
    axs[1, 0].set_ylim([0,1500])
    axs[1, 0].set_xlim([1,52])

    axs[1, 1].plot(tyden_85,hodnota_85, color="red")
    axs[1, 1].set_title('85+')
    axs[1, 1].set_ylim([0,1500])
    axs[1, 1].set_xlim([1,52])
    fig.show()
```

```python
make_plot(2020)
```
